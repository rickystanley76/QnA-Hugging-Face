```python
from transformers import pipeline
```

```python
question_answer = pipeline("question-answering")
```

```python
context ="Michael Jordan With five regular season MVP awards, a Defensive Player of the Year Award, six MVP finals, named to the All-Star team twelve times and named to the All-NBA First Team 10 times, there is no doubt that Michael Jordan has changed the game of basketball forever."
```

```python
result = question_answer(question= "How many times Michael Jordan became MVP?", context=context)
```

```python
print(f"Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}")
```
